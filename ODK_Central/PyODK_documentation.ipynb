## 1. What does PyODK do?


a. Provide a Python interface to interact with ODK Central.

b. Allows user authentication using API tokens and credentials.

c. Gelps list projects, forms, submissions, and attachments.

c. Supports downloading and uploading of submission data (including pandas dataframes)

d. Manage media files, form definintions and submision attachmeents.

e. Automating workflows; like syncing data into databases, runing analyses and generating reports.

## 1. Example 1: Authentication and fetching forms

```python
from pyodk.client import Client
```

```python
import pandas as pd
from pyodk.client import Client
```

```python
# Initialize Client using ~/.pyodk_config.toml
client = Client()

# Get a list of forms
forms = client.forms.list()
print(forms)
```

```python
if forms:
    # Get submissions for a form
    submissions = client.submissions.get_table(form_id='Image Safari Crop Scout (Phone Approach)')
    print(submissions)
else:
    print("⚠️ No forms found.")
```

```python
with Client() as client:
    client.open()
    print(client.projects.list())
```

```python
client.submissions.list(form_id='Image Safari Crop Scout (Phone Approach)')
```

```python
client.submissions.get_table(form_id='Image Safari Crop Scout (Phone Approach)')
```

```python
json = client.submissions.get_table(form_id='Image Safari Crop Scout (Phone Approach)') ['value']
import pandas as pd
df = pd.json_normalize(json, sep='++')
df.head(3)
```
